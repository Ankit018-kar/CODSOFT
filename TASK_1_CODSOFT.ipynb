```python
# TO DO LIST : TASK_1_CODSOFT

tasks=[]

def add_task():
    task=input("Enter the task : ")
    tasks.append(task)
    print(f"Task '{task}' is added to the list.")

def list_tasks():
    if not tasks:
        print("The task list is empty...")
    else:
        for index , task in enumerate(tasks):
            print(f" {index}) {task}")

def delete_tasks():
    list_tasks()
    task_to_delete = int(input("enter the index of the task to be deleted : "))
    try:
        if 0 <= task_to_delete < len(tasks):
            removed_task =tasks.pop(task_to_delete)
            print(f"Task : '{removed_task} has been deleted .")
        else:
            print(f"Task : '{task_to_delete} not found .")
    except ValueError:
        print("INVALID INPUT !!")

if __name__ == "__main__":
    print("Welcome to the to do list app !!")
    while True:
        print('''
                What do you want to do :
                1. ADD a task .
                2. DELETE a task .
                3. VIEW tasks .
                4. EXIT the app .    ''')

        choice = int(input("Enter the operation you want to perform :"))

        if choice == 1 :
            add_task()
        elif choice == 2 :
            delete_tasks()
        elif choice == 3 :
            list_tasks()
        elif choice == 4 :
            break
        else:
            print("Invalid Input Try Again !!")

print("Thanks for using the app !! ")
```
